```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
```

```python
gme = pd.read_csv(r'D:\mlprep\gme.csv', index_col= 'Date')
```

```python
gme.head()
```

```python
gme = gme[['Close']]
```

```python
gme.head()
```

```python
gme['moving_avg_7'] = gme['Close'].rolling(window=7).mean()
```

```python
plt.title('GME - 7-trading days Moving Average')
plt.plot(gme.index, gme['moving_avg_7'])
plt.xlabel('Trading Dates')
plt.ylabel('Price (USD)')
plt.xticks(gme.index[::12], rotation='vertical')
plt.legend(('$'), loc='upper left')
plt.show()
```

```python
alt.Chart(gme.reset_index()).mark_line().encode(
    
    x= alt.X('Date', title="Date", axis=alt.Axis(labelAngle=90)),
    y = alt.Y('moving_avg_7', title="Price (USD")
).properties(
    width=600,
    height=300,
    title='GME - 7-trading days Moving Average'
)
```
